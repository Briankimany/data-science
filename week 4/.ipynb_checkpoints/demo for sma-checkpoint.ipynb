```python
data = pd.read_csv("station_belem.csv")
data.head()
data_t = (data.set_index("YEAR")).T
data_2 = data.copy()

data_2['YEAR'] = (pd.to_datetime(data_2['YEAR'] , format = '%Y'))
data_2 = data_2.drop(columns = ['D-J-F', 'M-A-M', 'J-J-A', 'S-O-N', 'metANN'] , axis = 1 )


data_2['SMA'] = data_2.mean(axis = 1)


for index ,sma in enumerate(data_2['SMA']):
    if int(sma) > 50:
        data_2 = data_2.drop(index = index , axis = 0)

data_2['YEAR'] = pd.to_datetime(data_2['YEAR'] , format = '%Y').dt.year
data_2['YEAR'] = data_2['YEAR'].astype(str)
data_2.reset_index(inplace = True)


#getting the sma for diffrent periods

for i in range (2 ,6):
    data_2[f'SMA_{i}'] = data_2.SMA.rolling(i, min_periods=1).mean()

data_2.describe()
```

```python
colors = ['green', 'red', 'orange' , 'blue' , 'green']
# Line plot
data_2[['YEAR', 'JAN','SMA', 'SMA_2', 'SMA_3', 'SMA_4', 'SMA_5']].plot(x='YEAR', y=['SMA_5' , 'JAN'] ,color = colors )
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(labels =['SMA', 'SMA_2', 'SMA_5', 'SMA_3', 'SMA_4'], fontsize=14)
plt.title('plot', fontsize=20)
plt.xlabel('YEAR', fontsize=16)
plt.ylabel('Temperature [Â°C]', fontsize=16)
```
