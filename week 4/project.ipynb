```python
import matplotlib.pyplot as plt
import pandas as pd
def get_():
    l = []
    while True :
        num = (input()).split(' ')
        if num[0] == 'b':
            return l
            break

        else:
            for i in num:
                #print (i)
                try:
                    num = float(i)
                    l.append(num)
                    #print ("this is l",l)
                except Exception as e:
                    pass

    return l

nums  = get_()

data_dict = {}
degrees = 0
def get_x(nums , angle_change = 10):
    c = 7.1
    values = []
    angles = []
    angle = 0
    for index , num in enumerate ( nums) :
        value =( num - c)
        values.append(value)
        angles.append(angle)
        angle +=angle_change

    return values , angles

##getting the x change
delta_x , angle_changes = get_x(nums , angle_change = 10)

data_dict['angle'] = angle_changes
data_dict['y_values'] = delta_x

df = pd.DataFrame.from_dict(data_dict)




#### by apandas

df['angle'] =df['angle'].astype(int)
df['y_values'] =df['y_values'].astype(int)
df.plot(kind = 'line' , x = 'angle' , y = 'y_values')



# ## by pyplot
# x = df['angle']
# y = df['y_values']

# # Create a basic line plot
# plt.plot(x, y)
# plt.xlabel('X-axis label')
# plt.ylabel('Y-axis label')
# plt.title('Title of the Plot')
```
