```python
# Creating a dictionary of months with respective month numbers
months = {
    "January": 1,
    "February": 2,
    "March": 3,
    "April": 4,
    "May": 5,
    "June": 6,
    "July": 7,
    "August": 8,
    "September": 9,
    "October": 10,
    "November": 11,
    "December": 12
}
import  random as rd

g = [i for i in range (7)]
# Creating an imaginary number of car sales for the next two years with an increasing trend
car_sales = {
    "2023": {month: (index + 1) * 1000  if index < 10 and index >= 3  else  (rd.choice(g) + 1) * 1000   for index, month in enumerate(months)},
    "2024": {month: (index + 2) * 1000  if index < 10 and index >= 3  else  (rd.choice(g) + 1) * 1000   for index, month in enumerate(months) },
    "2025": {month: (index + 3) * 1000  if index < 10 and index >= 3  else  (rd.choice(g) + 7) * 1000   for index, month in enumerate(months) },
    "2026": {month: (index + 4) * 1000  if index < 10 and index >= 3  else  (rd.choice(g) + 9) * 1000   for index, month in enumerate(months) },
    "2027": {month: (index + 5) * 1000  if index < 10 and index >= 3  else  (rd.choice(g) + 2) * 1000   for index, month in enumerate(months) },
    "2028": {month: (index + 6) * 1000  if index < 10 and index >= 3  else  (rd.choice(g) + 4) * 1000   for index, month in enumerate(months) },
    "2029": {month: (index + 7) * 1000  if index < 10 and index >= 3  else  (rd.choice(g) + 6) * 1000   for index, month in enumerate(months) },
    "2030": {month: (index + 8) * 1000  if index < 10 and index >= 3  else  (rd.choice(g) + 8) * 1000   for index, month in enumerate(months) },
    "2031": {month: (index + 9) * 1000  if index < 10 and index >= 3  else  (rd.choice(g) + 4) * 1000   for index, month in enumerate(months) },
    "2032": {month: (index + 10) * 1000  if index < 10 and index >= 3  else  (rd.choice(g) + 3) * 1000   for index, month in enumerate(months) },
    "2033": {month: (index + 11) * 1000  if index < 10 and index >= 3  else  (rd.choice(g) + 7) * 1000   for index, month in enumerate(months) }
}
d_2023 = car_sales['2023']

A = pd.DataFrame.from_dict(d_2023 , orient= 'index' , columns = ['2023'])
A['time'] = A.index
n = ['time' , '2023']
A = A[n]
for index , year in enumerate(car_sales):
    if index !=0:
        d_2023 = car_sales[year]
        d = pd.DataFrame.from_dict(d_2023 , orient= 'index' , columns = [f'{year}'])
        d['time'] = d.index
        A= pd.merge(A , d , on ='time' , how = 'right')
        
A
```
